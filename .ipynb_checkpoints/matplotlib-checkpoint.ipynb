```python
import matplotlib.pyplot as plt
```

```python
x=[2,3,4,5,6]
y1=[4,5,6,7,8]
plt.plot(x,y1)
plt.show()
```

```python
x=[2,3,4,5,6]
y1=[4,5,6,7,8]
y2=[6,7,8,9,10]

plt.subplot(2,2,1)
plt.plot(x,y1)

plt.subplot(2,2,2)
plt.plot(x,y2)
plt.show()
```

```python
x=[2,3,4,5,6]
y1=[4,5,6,7,8]
y2=[6,7,8,9,10]
y3=[10,11,12,13,14]
y4=[13,14,15,16,17]

plt.suptitle("Data")
plt.subplot(2,2,1)
plt.title("Graph-1")
plt.plot(x,y1)

plt.subplot(2,2,2)
plt.title("Graph-2")
plt.plot(x,y2)

plt.subplot(2,2,3)
plt.title("Graph-3")
plt.plot(x,y3)

plt.subplot(2,2,4)
plt.title("Graph-4")
plt.plot(x,y4)
plt.show()
```
