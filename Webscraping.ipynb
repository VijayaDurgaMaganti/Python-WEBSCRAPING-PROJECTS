```python
prop=soup.find_all('div',attrs={"class":"title-line"})
```

```python
prop_name=[]
for i in prop:
    prop_name.append(i.text)
```

```python
prop_name=[i.text for i in prop]
print(prop_name)
print(len(prop_name))
```

```python
b=soup.find_all('span',attrs={"itemprop":"addressLocality"})
location=[i.text for i in b]
print(location)
print(len(location))
    
```

```python
c=soup.find_all('div',attrs={"data-type":"price-link"})
price=[i.text for i in c]
print(price)
print(len(price))
```

```python
d=soup.find_all('td',attrs={"class":"lbl rate"})
sqft=[i.text for i in d]
print(sqft)
print(len(sqft))
```

```python
e=soup.find_all('td',attrs={"class":"val"})
val=[i.text for i in e]
print(val)
print(len(val))
```

```python
f=soup.find_all('td',attrs={"class":"size"})
area=[i.text for i in f]
print(area)
print(len(area))
```
